### Step 1

```python
import xlrd

book = xlrd.open_workbook('house_price.xls')

sheet = book.sheet_by_name('price')

price_list = []
bath_list = []
htype_list = []
lot_size_list = []
year_list = []
area_list = []

for i in range(sheet.nrows):
    price,bath,htype,lot_size,year,area = sheet.row_values(i)
    row = sheet.row_values(i)
    if i !=0:
     ##   print(price,bath,htype,lot_size,year,area)
        price_list.append(price)
        bath_list.append(bath)
        htype_list.append(htype)
        lot_size_list.append(lot_size)
        year_list.append(year)
        area_list.append(area)
print(price_list)
print(bath_list)
print(htype_list)
print(lot_size_list)
print(year_list)
print(area_list)
```

```python
import matplotlib.pyplot as plt
```

### Step 2

```python
plt.plot(year_list,price_list)
plt.xlabel('year')
plt.ylabel('price')
plt.show()
```

### Step 3

```python
plt.hist(price_list)
plt.show()
```

```python
point_size = [i**3 for i in bath_list]
plt.xlabel('house area')
plt.ylabel('house price')
plt.scatter(area_list, price_list, s=point_size)
plt.show()
```
